### 카카오 API로 주소의 위도/경도 탐색

#### 탐색하지 못하는 주소가 있으면 점진적으로 주소를 단순화시킬 예정

```python
print(ev_charger['주소'].loc[50])
print('↓')
print(ev_charger['주소'].loc[50].split(',')[0].split('(')[0])
```

```python
GCS = []

def get_location(address):
  url = 'https://dapi.kakao.com/v2/local/search/address.json?query=' + address
  headers = {"Authorization": "KakaoAK 3c768412ed253db7abaf7e0d5a15788f"}
  api_json = json.loads(str(requests.get(url,headers=headers).text))
  address = api_json['documents'][0]['address']
  crd = {"lat": str(address['y']), "lng": str(address['x'])}
  address_name = address['address_name']

  return crd

for i, addr in enumerate(ev_charger['주소']):
    addr = ev_charger['주소'].loc[i].split(',')[0].split('(')[0]
    while(1):
        try:
            crd = get_location(addr)
            #print(i, crd['lat'], crd['lng'])
            break
        except:
            addr = addr[:-1]

    GCS.append((crd['lat'], crd['lng']))
```

```python
print('가로길이', poly_right-poly_left)
print('세로길이', poly_up-poly_down)
print('세종시는', int((poly_right-poly_left)/100), '*', int((poly_up-poly_down)/100), '의 격자로 표현 가능 (100m x 100m)')
```

### 위경도(WGS84 좌표계)를 GRS80 좌표계로 변환

```python
def WGS84_to_GRS80(WGS):
    GRS = []

    # 한 격자당 위도 변화
    length_100m = (36.73376199366808 - 36.40675835257603) / 364

    # 한 격자당 경도 변화
    width_100m = (127.4092754044858 - 127.12765872447403) / 254

    for i, w in enumerate(WGS):
        GRS_x = 966700 + (float(w[1]) - 127.12765872447403)/width_100m * 100
        GRS_y = 1823200 + (float(w[0]) - 36.40675835257603)/length_100m * 100
        GRS.append((GRS_x, GRS_y))
        
    return GRS
```

```python
WGS = ev_charger['WGS84']
GRS = WGS84_to_GRS80(WGS)

ev_charger['GRS80'] = GRS
ev_charger.head()
```

#### 위치별 딕셔너리를 생성해서 주용도별로 분류 후, 리스트에 정리

```python
reg = []

for i in building_reg.drop_duplicates(['GRS80']).index:
    dic = {'교육연구시설':0, '주거지':0, '농축산업시설':0, '창고시설':0, '편의시설':0, '생산시설':0, '님비시설':0,
       '자동차관련시설':0}
    temp = building_reg[building_reg['GRS80']==building_reg['GRS80'].loc[i]]
    for j in range(len(temp['주용도코드명'].unique())):
        dic[temp['주용도코드명'].unique()[j]] = temp['주용도코드명'].value_counts()[j]
    reg.append(dic)
```

```python
reg[:2]
```

#### 중복을 제거한 폴리곤 좌표

```python
building_reg_p = building_reg[['GRS80']].drop_duplicates(['GRS80'])
building_reg_p.head()
```

#### 주용도별로 열 생성

```python
for n in building_reg['주용도코드명'].unique():
    temp = []
    for r in reg:
        temp.append(r[n])
    building_reg_p[n] = temp

building_reg_p.head()
```

```python
building_reg_p.reset_index(drop=False, inplace=True)
```

#### 데이터 병합 ( 격자 내에 존재하는 지 확인하며 count )

```python
geometry = material.geometry.copy().astype('string')
```

### 격자별 읍면동 주소 찾기

```python
geo = geopandas.read_file('./dataset/전처리전/전국_읍면동_경계.geojson', encoding='utf-8')
geo = geo[geo['sidonm'] == '세종특별자치시']
```

```python
temp = []
for mp in geo['geometry'].astype(str):
    temp.append(shapely.wkt.loads('Polygon ((' + mp.split('(((')[1].split(')))')[0] + '))'))

address = []
for i in range(len(material)):
    check = 0
    cnt = 0
    lat = material['lat'].loc[i]
    lng = material['lng'].loc[i]
    
    for j, t in enumerate(temp):
        if Point(lng, lat).within(t):
            address.append(geo.loc[j]['temp'].split()[1])
            check = 1
            break
            
    if check == 0:
        brake = 0
        while 1:
            lat_temp = material['lat'].loc[i]
            lng_temp = material['lng'].loc[i]
            
            if cnt % 4 == 0:
                lng_temp += 0.001 * (cnt//4 + 1)
                lat_temp += 0.001 * (cnt//4 + 1)
            elif cnt % 4 == 1:
                lng_temp += 0.001 * (cnt//4 + 1)
                lat_temp -= 0.001 * (cnt//4 + 1)
            elif cnt % 4 == 2:
                lng_temp -= 0.001 * (cnt//4 + 1)
                lat_temp += 0.001 * (cnt//4 + 1)
            elif cnt % 4 == 3:
                lng_temp -= 0.001 * (cnt//4 + 1)
                lat_temp -= 0.001 * (cnt//4 + 1)
                
            cnt += 1
            
            for j, t in enumerate(temp):
                if Point(lng_temp, lat_temp).within(t):
                    address.append(geo.loc[j]['temp'].split()[1])
                    brake = 1
                    break
                    
            if brake == 1:
                break
            
len(address)
```
